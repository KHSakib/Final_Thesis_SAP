```python
import os
import cv2

folder_path = "/content/drive/MyDrive/Data set/LD-Dataset-Full/true"
output_folder = "/content/drive/MyDrive/Data set/LD-Dataset-Full-Frames/true"
frame_increment = 50  # Assuming you want to capture every frame

# Ensure the output folder exists, if not, create it
os.makedirs(output_folder, exist_ok=True)

# Iterate over files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.mp4'):
        print(filename)
        video_path = os.path.join(folder_path, filename)

        # Open the video file
        cam = cv2.VideoCapture(video_path)
        current_frame = 0

        while True:
            # Skip frames based on frame_increment
            for _ in range(frame_increment):
                ret, frame = cam.read()

            if ret:
                # Write the frame as an image
                image_name = f'{output_folder}/{filename}_{current_frame}.jpg'
                print(f'Creating... {image_name}')
                cv2.imwrite(image_name, frame)
                current_frame += frame_increment
            else:
                break

        # Release video capture object
        cam.release()
        cv2.destroyAllWindows()
```
